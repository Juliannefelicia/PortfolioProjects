```python
#Import Libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjust the configuration of the plots we will create

df = pd.read_csv(r"C:\Users\User\Downloads\movies.csv")
```

```python
#Looking at the data
df.head()
```

```python
# Check for missing data
for col in df.columns:
    pctMissing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pctMissing))
```

```python
df = df.dropna()
for col in df.columns:
    pctMissing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pctMissing))
```

```python
# Data types for columns
df.dtypes
```

```python
#change data type
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
#Create correct year
df['yearcorrect'] = df['released'].astype(str).str.split().str[2]
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Drop any duplicates
df.drop_duplicates().head()
```

```python
# Budget high correlation
# Company high correlation
```

```python
# Scatter plot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross regression using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
# Let's start looking at correlation, can simply use df.corr()
df.corr(method='pearson')
#Pearson (default), kendall, spearman
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Look at company
df.head()
```

```python
# Numerize companies
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
